## 2.0 Module to recreate the dataframe

The current dataframe contains important information (`contents`) in `list` format. We split the list to different columns, and the column names are different depending on the `eventType`.

1. Change the type of `regDate` from string to datetime.
2. Split three elements of `contents` entry, save them separately, and drop `contents` column.
3. Convert the type of contents from string to integer.

```python
def convertContents2Float(contents):
    try:
        return float(contents)
    except:
        if contents == 'null':
            return np.nan
        else:
            return contents

def change2df(eventType, phone_number=None):
    if phone_number:
        target_df = sensor_df[(sensor_df.eventType == eventType) & (sensor_df.phone == phone_number)]
    else:
        target_df = sensor_df[sensor_df.eventType == eventType]
    target_df.reset_index(inplace=True)
    target_df.regDate = target_df.regDate.map(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f'))

    if eventType == 8:
        converted_columns = ['contentsRoom', 'contentsCurrAct', 'contentsPrevAct']
    elif eventType == 73:
        converted_columns = ['contentsStep', 'contentsDistance', 'contentsCalory',
                               'contentsHrm', 'contentsBattery',
                               'contentsCondition', 'contentsTotalSleep',
                               'contentsTotalDeepSleep', 'contentsTotalLightSleep']
    else:
        raise ValueError("Eventtype {} is not available".format(eventType))
        
    temp_df = pd.DataFrame(target_df.contents.map(lambda x: x.split(';')).values.tolist(), columns=converted_columns)
    for column in temp_df:
        temp_df[column] = temp_df[column].map(lambda x: convertContents2Float(x))
    target_df = pd.concat([target_df, temp_df], axis=1)
    target_df.drop(columns=['contents'], inplace=True)
    target_df = target_df[['id', 'phone', 'devType', 'macAddr', 'eventType'] + converted_columns +\
                         ['commState', 'powerState', 'batState', 'regDate', 'createdAt', 'updatedAt']]
    
    if eventType == 8:
        target_df.contentsRoom = target_df.contentsRoom.astype('category').cat.codes
   
    return target_df
```

### 2.1.2 contentsCurrAct vs contentsPrevAct <font color=blue>(PROBLEM DETECTED)</font>

According to the official document, `contentsCurrAct` is the activity status at the current location, and `contentsPrevAct` is the activity status at the previous location. It means that <font color=red>`contentsPrevAct` data must be `contentsCurrAct` shifted by one activity datapoint,</font> <font color=blue size=5>which is not the case.</font> The following plots show the plots of `contentsCurrAct` and `contentsPrevAct`:

```python
plt.figure(figsize=(16, 8))
plt.plot(user_01220461115.createdAt, user_01220461115.contentsCurrAct, label='Curr activity')
plt.plot(user_01220461115.createdAt, user_01220461115.contentsPrevAct, label='Prev activity')
plt.xlabel('Date', fontdict=fontLabel)
plt.ylabel('Activity', fontdict=fontLabel)
plt.legend()
plt.savefig('./FigureCompilation/user_01220461115_activity_per_time.png')
```

Those two are completely different data with no whatsoever relationship.

I suspected it could be because the data are ordered by wrong time data. There are three different time data in the dataset:
1. `regDate`: <font color=red size=5>Detailed description needed</font>
2. `createdAt`: <font color=red size=5>Detailed description needed</font>
3. `updatedAt`: <font color=red size=5>Detailed description needed</font>

So I re-ordered data by different time data, and investigated if `CurrAct` matches `PrevAct` by any chance. It turns out that <font color=blue size=5>`PrevAct` a time step does not match the `CurrAct` of the previous timestep in all cases, whether or not you sort the data by `regDate`, `createdAt`, or `updatedAt`.</font>

```python
# Investigate the consistency between `PrevAct` at the current timestep and `CurrAct` at the previous timestep.
# `eventType==8` has `PrevAct` and `CurrAct`

def sort_and_investigate(target_user_phone, sort_by=None):
    sorted_list = []
    target_user_df = change2df(8, target_user_phone)
    if sort_by:
        target_user_df = target_user_df.sort_values(by=sort_by)
        target_user_df.reset_index(inplace=True, drop='index')
    for index, row in target_user_df.iterrows():
        if index == 0:
            prev_prev_act = row.contentsCurrAct
            continue
        else:
            curr_prev_act = row.contentsPrevAct
            if curr_prev_act != prev_prev_act:
                sorted_list.append(index)
            prev_prev_act = row.contentsCurrAct
    return sorted_list
```

```python
without_sort = sort_and_investigate('01220461115')
print("Percentage of inconsistent data:", without_sort.__len__()/change2df(8, '01220461115').__len__())
```

This raises another problem in the time data: <font color=blue size=5>the relationship between `regDate`, `createdAt`, and `updatedAt` is different by some users.</font>

The following code examines which users show different relationships between the time data:

```python
import itertools
from threading import Thread

validity_check_list = []
def checkTimeDataConsistency(user):
    global validity_check_list
    without_sort = sort_and_investigate(user)
    with_sort_regDate = sort_and_investigate(user, 'regDate')
    with_sort_createdAt = sort_and_investigate(user, 'createdAt')
    with_sort_updatedAt = sort_and_investigate(user, 'updatedAt')
    
    for i, j in itertools.combinations([without_sort, with_sort_createdAt, with_sort_updatedAt, with_sort_regDate], 2):
        if i != j:
            validity_check_list.append(user)
            break
#     print(user, without_sort == with_sort_createdAt, without_sort == with_sort_updatedAt, without_sort == with_sort_regDate)
    
# with cf.ThreadPoolExecutor(max_workers=1000) as executor:
#     executor.map(checkTimeDataConsistency, users_eventType8)

for user in users_eventType8:
    t = Thread(target=checkTimeDataConsistency, args=(user,))
    t.start()
t.join()
```

```python
validity_check_list.__len__()
```

```python
set(validity_check_list).__len__()
```

```python
print("Percentage of users with inconsistent time data:", set(validity_check_list).__len__()/set(sensor_df.phone).__len__())
```

For the rest of users, the time data are all in the same order.

```python
for user in set(sensor_df.phone):
    if user not in validity_check_list:
        without_sort = sort_and_investigate(user)
        with_sort_regDate = sort_and_investigate(user, 'regDate')
        with_sort_createdAt = sort_and_investigate(user, 'createdAt')
        with_sort_updatedAt = sort_and_investigate(user, 'updatedAt')
        inconsistent_time_data = 0
        for i, j in itertools.combinations([without_sort, with_sort_createdAt, with_sort_updatedAt, with_sort_regDate], 2):
            if i != j:
                print(user)
                inconsistent_time_data += 1
        if inconsistent_time_data == 0:
            print("User:", user, "without_sort == regDate:", (without_sort == with_sort_regDate), "\t",
                  "regDate == createdAt:", (with_sort_regDate == with_sort_createdAt), "\t",
                  "createdAt == updatedAt:", (with_sort_createdAt == with_sort_updatedAt), "\t")
```

# Make analysis model into a module

```python
class sensor_data_analysis:
    def __init__(self, eventType, phone_number=None):
        self.eventType = eventType
        if eventType != 8 and eventType != 73:
            raise ValueError('Unknown eventType')

        self.phone_number = phone_number
        if phone_number:
            self.sensor_df = pd.read_sql_query('select * from SensorData where eventType = {} and phone = {}'.format(eventType, phone_number), mysql)
        else:
            self.sensor_df = pd.read_sql_query('select * from SensorData where eventType = {}'.format(eventType), mysql)

        self.target_df = self.change2df()



    def convertContents2Float(self, contents):
        try:
            return float(contents)
        except:
            if contents == 'null':
                return np.nan
            else:
                return contents



    def change2df(self):
        if self.phone_number:
            target_df = self.sensor_df[(self.sensor_df.eventType == self.eventType) & (self.sensor_df.phone == self.phone_number)]
        else:
            target_df = self.sensor_df[self.sensor_df.eventType == self.eventType]
        target_df.reset_index(inplace=True)
        target_df.regDate = target_df.regDate.map(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f'))

        if self.eventType == 8:
            converted_columns = ['contentsRoom', 'contentsCurrAct', 'contentsPrevAct']
        elif self.eventType == 73:
            converted_columns = ['contentsStep', 'contentsDistance', 'contentsCalory',
                                   'contentsHrm', 'contentsBattery',
                                   'contentsCondition', 'contentsTotalSleep',
                                   'contentsTotalDeepSleep', 'contentsTotalLightSleep']
        else:
            raise ValueError("Eventtype {} is not available".format(self.eventType))
            
        temp_df = pd.DataFrame(target_df.contents.map(lambda x: x.split(';')).values.tolist(), columns=converted_columns)
        for column in temp_df:
            temp_df[column] = temp_df[column].map(lambda x: convertContents2Float(x))
        target_df = pd.concat([target_df, temp_df], axis=1)
        target_df.drop(columns=['contents'], inplace=True)
        target_df = target_df[['id', 'phone', 'devType', 'macAddr', 'eventType'] + converted_columns +                         ['commState', 'powerState', 'batState', 'regDate', 'createdAt', 'updatedAt']]
        if self.eventType == 8:
            target_df.contentsRoom = target_df.contentsRoom.astype('category').cat.codes
       
        return target_df



    def getNormalizedResult(self, analysis_content, period):
        if period != 'regDate':
            self.target_df['hour'] = self.target_df.createdAt.map(lambda x: datetime.strftime(x, '%H'))
            self.target_df['dayHour'] = self.target_df.createdAt.map(lambda x: str(x.weekday()) + datetime.strftime(x, '%H'))
        content_time_data = self.target_df[[period, analysis_content]]
        mean_content_time_data = content_time_data.groupby(period).mean()
        normalized_mean_content_time_data = mean_content_time_data/max(mean_content_time_data.values)
        if period != 'regDate':
            std_content_time_data = content_time_data.groupby(period).std()
            std_content_time_data = std_content_time_data/max(std_content_time_data.values)
        else:
            std_content_time_data = np.nan
        
        # Insert empty time slots with `np.nan`
        idx_list = normalized_mean_content_time_data.index
        if period == 'hour':
            time_list = range(24)
        else:
            time_list = []
            for j in range(7):
                time_list.extend([str(j) + str(i).zfill(2) for i in range(24)])
            
        if period != 'regDate':
            for time in time_list:
                zfill_time = str(time).zfill(2)
                if zfill_time not in idx_list:
                    normalized_mean_content_time_data.loc[zfill_time] = np.nan
                    std_content_time_data.loc[zfill_time] = np.nan
            normalized_mean_content_time_data = normalized_mean_content_time_data.sort_index()
            std_content_time_data = std_content_time_data.sort_index()
            
        return normalized_mean_content_time_data, std_content_time_data



    def plotContent(self, analysis_content_list, period, std=True):
        plt.figure(figsize=(16,8))
        for analysis_content in analysis_content_list:
            mean_content_time_data, std_content_time_data = self.getNormalizedResult(analysis_content, period)
            ymin = (mean_content_time_data-std_content_time_data)[analysis_content]
            ymax = (mean_content_time_data+std_content_time_data)[analysis_content]
            plt.plot(mean_content_time_data.index, mean_content_time_data, 'o-', label=analysis_content.split('contents')[1])
            if std:
                plt.fill_between(mean_content_time_data.index, ymin, ymax, alpha=0.5)
        if period == 'dayHour':
            plt.xticks([str(i)+'00' for i in range(7)], ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
            # plt.vlines([str(i)+'00' for i in range(7)], ymin=min(ymin)/3,
            #           ymax=max(ymax)/3, color='red')
        plt.xlabel("Hour")
        plt.ylabel("Normalized Activity")
        plt.legend()
```
